```python
# Reading the data:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn


df = pd.read_csv(r"C:\Users\LENOVO\Downloads\FoodBalanceSheets_E_Africa_NOFLAG (1).csv", encoding="latin-1")
df.head()
```

```python
# What is the mean and standard deviation across the whole dataset for the year
# 2017 to 2 decimal places?

# Calculate the mean of column:
mean = df['Y2017'].mean()

# Calculate the standard deviation of column:
std = df['Y2017'].std()

# Round the mean and standard deviation to 2 decimal places:
mean = round(mean, 2)
std = round(std, 2)

# Print the results
print("Mean of column 'Y2017':", mean)
print("Standard deviation of column 'Y2017':", std)
```

```python
# What is the total number and percentage of missing data in 2014 to 3 decimal
# places?
# Count the number of missing values:

df.Y2014.isnull().sum()

# Total number of values in the DataFrame:

missing_values = df.Y2014.isnull().sum().sum()
total_values = df.Y2014.size
percentage = round(missing_values / total_values * 100, 3)
print("Number of missing values:", missing_values)
print("Percentage of missing values:", percentage, "%")
```
